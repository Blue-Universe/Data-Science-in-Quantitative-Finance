```python
import pandas as pd
import re
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://classcodes.com/naics-code-list/'
```

```python
from fake_useragent import UserAgent
ua = UserAgent()
```

```python
header = {'User-Agent':str(ua.chrome)}
htmlContent = requests.get(url, headers=header)
```

```python
NAICS_data = pd.DataFrame(columns=['NAICS Code', 'NAICS Title'])
match = re.finditer(r'NAICS(&nbsp;| )+Code(&nbsp;| )+\d{6}(&nbsp;| )+-(&nbsp;| )+.*?&', htmlContent.text)
for i in match:
    item = i.group()
    item = item.replace('&nbsp;',' ')
    code = re.search(r'\d{6}',item).group(0)
    name = '-'.join(item.split('-')[1:])[1:-1]
    NAICS_data = NAICS_data.append([{'NAICS Code':int(code), 'NAICS Title':name}],ignore_index=True)
```
