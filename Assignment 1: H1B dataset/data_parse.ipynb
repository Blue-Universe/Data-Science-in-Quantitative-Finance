```python
def tofloat(x):
    try:
        return float(x)
    except:
        return np.nan
```

```python
def cleantype(x,type0 = str):
    if pd.isnull(x):
        return x
    else:
        try:
            return type0(x)
        except:
            return np.nan

def cleandata(df):
    for i in range(df.shape[1]):
        if len(df.iloc[:,i].dropna().apply(type).value_counts())!=1 and len(df.iloc[:,i].dropna().apply(type).value_counts())!=0:
            type0 = df.iloc[:,i].dropna().apply(type).value_counts().index[0]
            df.iloc[:,i] = df.iloc[:,i].apply(cleantype, type0 = type0)
    return df
```

```python
dict1 = {'Y': 1, 'M': 12, 'B': 26,'W':52,'H':2000}
def findrate(x):
    try:
        return dict1[x]
    except:
        return 1
dict2 = {'Year': 1, 'Month': 12, '2 weeks': 26,'Week':52,'Hour':2000}
def findrate1(x):
    try:
        return dict2[x]
    except:
        return 1
```

```python
def selectdata_20016(X):
    X = cleandata(X)
    data = pd.DataFrame(columns = head)
    data['date (of application)'] = X['CASE_SUBMITTED']
    data['case_number'] = X['CASE_NUMBER']
    data['employer'] = X['EMPLOYER_NAME']
    data['tot_foreign_positions'] = X['TOTAL_WORKERS']
    data['naic_code'] = X['NAIC_CODE']
    data['job_title'] = X['JOB_TITLE']
    #data['job_code'] = X['OCCUPATIONAL_CODE']
    data['soc_name'] = X['SOC_NAME']
    data['approval_status'] = X['CASE_STATUS']
    data['annual_wage'] = X['WAGE_RATE_OF_PAY_FROM']*X['WAGE_UNIT_OF_PAY'].apply(findrate1)
    data['annual_prevailing_wage'] = X['PREVAILING_WAGE']*X['WAGE_UNIT_OF_PAY'].apply(findrate1)
    data['city (of occupation)'] = X['WORKSITE_CITY']
    data['state (of occupation)'] = X['WORKSITE_STATE']
    data['county (of occupation)'] = X['WORKSITE_COUNTY']
    return data
data_16 = selectdata_20016(LAC_2016[LAC_2016['VISA_CLASS']=='H-1B'])
```
